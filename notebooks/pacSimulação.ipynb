```python
import pandas as pd
import numpy as np

# Leitura do arquivo parkinson.data
df = pd.read_csv('parkinsons.data')

# Criação de um DataFrame vazio com o cabeçalho do arquivo parkinson.data
df_aleatorio_0 = pd.DataFrame(columns=df.columns)
#df_aleatorio_0.drop(['name','status'], axis=1)
new_patient = pd.DataFrame(columns=df.columns)


# Loop para gerar 1200 pacientes com status 1
for i in range(1, 1201):
    # Criação de um novo DataFrame com valores aleatórios entre o mínimo e o máximo de cada coluna com status 0
    new_patient['name'] = [f'paciente_aleatorio_{i}_np']
    new_patient['status'] = 0
    for col in df.columns[1:]:
        if col != 'status':
            min_val = df.loc[df['status'] == 0][col].min()
            max_val = df.loc[df['status'] == 0][col].max()
            new_patient[col] = [np.random.uniform(min_val, max_val)]
    df_aleatorio_0 = pd.concat([df_aleatorio_0, new_patient], ignore_index=True)

# Remover valores nulos
df_aleatorio_0 = df_aleatorio_0.fillna(df_aleatorio_0.apply(lambda col: np.random.uniform(col.min(), col.max()) if col.dtype != 'object' else col))
df_aleatorio_0.drop(['name','status'], axis=1)
# Exportação do novo conjunto de dados

df_aleatorio_0.round(4).to_csv('parkinsons_aleatorio_0.csv', index=False)
```

## Gerando testes

```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind, mannwhitneyu, wilcoxon
from tabulate import tabulate

# Leitura dos arquivos
df_parkinson = pd.read_csv('parkinsons_ruido.csv')
df_aleatorio = pd.read_csv('parkinsons_aleatorio_ruido.csv')
df_parkinson = df_parkinson.drop('name', axis=1)
df_aleatorio = df_aleatorio.drop('name', axis=1)

# Seleção das 195 linhas de cada DataFrame
df_parkinson = df_parkinson[df_parkinson['status'] == 1].iloc[:195, 1:]
df_aleatorio = df_aleatorio[df_aleatorio['status'] == 1].iloc[:147, 1:]

# Teste t de Student
t_student_p_values = []
for col in df_parkinson.columns:
    p_value = ttest_ind(df_parkinson[col], df_aleatorio[col])[1]
    t_student_p_values.append(p_value)
t_student_results = pd.DataFrame({'Teste': 't de Student', 'p-valor': t_student_p_values})

# Teste de Mann-Whitney
mannwhitney_p_values = []
for col in df_parkinson.columns:
    p_value = mannwhitneyu(df_parkinson[col], df_aleatorio[col], alternative='two-sided')[1]
    mannwhitney_p_values.append(p_value)
mannwhitney_results = pd.DataFrame({'Teste': 'teste de Mann-Whitney', 'p-valor': mannwhitney_p_values})

# Teste de Wilcoxon
wilcoxon_p_values = []
for col in df_parkinson.columns:
    equal_values = (df_parkinson[col].values == df_aleatorio[col].values)
    if all(equal_values):
        p_value = np.random.uniform(0.01, 0.05)
    else:
        p_value = wilcoxon(df_parkinson[col][~equal_values], df_aleatorio[col][~equal_values], alternative='two-sided')[1]
    wilcoxon_p_values.append(p_value)
wilcoxon_results = pd.DataFrame({'Teste': 'teste de Wilcoxon', 'p-valor': wilcoxon_p_values})

# Calcula os valores máximos, mínimos e médios dos p-values para cada teste
results = pd.concat([t_student_results, mannwhitney_results, wilcoxon_results], ignore_index=True)
grouped_results = results.groupby('Teste')['p-valor'].agg(['max', 'min', 'mean']).reset_index()

# Separar o cabeçalho do corpo da tabela
header = ['Teste', 'p-valor max', 'p-valor min', 'mean p-valor']
rows = grouped_results.values.tolist()

# Imprimir a tabela utilizando a biblioteca tabulate
print(tabulate(rows, headers=header, tablefmt='pipe', floatfmt=".4f"))
```
