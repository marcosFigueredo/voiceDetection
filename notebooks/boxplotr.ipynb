```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar os dados
data_path = 'parkinsons.data'
df = pd.read_csv(data_path)

# Configurações para a visualização
sns.set(style="whitegrid")

# Selecionar características para os boxplots
features = ['MDVP:Fo(Hz)', 'MDVP:Jitter(%)', 'MDVP:Shimmer']

# Criar figura e eixos para os subplots
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(18, 6))

# Título geral
fig.suptitle('Distribuição de Características da Voz por Status', fontsize=16)

# Gerar um boxplot para cada característica selecionada
for i, feature in enumerate(features):
    sns.boxplot(x='status', y=feature, data=df, ax=axes[i], palette="Set2")
    axes[i].set_title(feature)
    axes[i].set_xlabel('Status')
    axes[i].set_ylabel('')
    axes[i].set_xticklabels(['Saudável (0)', 'PD (1)'])

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Ajusta o layout para o título geral
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
data_path = 'parkinsons.data'
df = pd.read_csv(data_path)

# Visualization settings
sns.set(style="whitegrid")

# Select features for boxplots
features = ['MDVP:Fo(Hz)', 'MDVP:Jitter(%)', 'MDVP:Shimmer']

# Create figure and axes for subplots
fig, axes = plt.subplots(nrows=1, ncols=len(features), figsize=(18, 6), dpi=300)

# General title
fig.suptitle('Distribution of Voice Features by Status', fontsize=18, fontweight='bold')

# Generate a boxplot for each selected feature
for i, feature in enumerate(features):
    sns.boxplot(x='status', y=feature, data=df, ax=axes[i], palette="Set2")
    axes[i].set_title(feature, fontsize=16, fontweight='bold')
    axes[i].set_xlabel('Status', fontsize=13, fontweight='bold')
    axes[i].set_ylabel('', fontsize=13, fontweight='bold')
    axes[i].set_xticklabels(['Healthy', 'PD'], fontsize=13, fontweight='bold')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout for the general title
plt.show()
```
