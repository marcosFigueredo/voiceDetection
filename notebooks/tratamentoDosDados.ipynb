```python
# Tratamento dos dados
import pandas as pd
import numpy as np
from tabulate import tabulate

# Carrega o arquivo CSV
df = pd.read_csv('parkinsons_aleatorio_1.csv')


# Normaliza os dados de todas as linhas
df_1_norm = (df - df.min()) / (df.max() - df.min())

# Verifica se existe algum dado NAN ou 0 e substitui por um valor aleatório entre o máximo e mínimo da coluna
for col in df_1_norm.columns:
    min_val = df_1_norm[col].min()
    max_val = df_1_norm[col].max()
    df_1_norm[col] = df_1_norm[col].apply(lambda x: np.random.uniform(min_val, max_val) if pd.isnull(x) or x == 0 else x)
    

# Salva os arquivos
df_1_norm.to_csv('parkinsons_norm_1.csv', index=False)

# Sumário dos dados
header = ['Status', 'Mínimo', 'Máximo', 'Média', 'Desvio Padrão']
rows = []
for df_norm, status in zip([df_1_norm, df_1_norm], [1, 0]):
    for col in df_norm.columns:
        rows.append([status, df_norm[col].min(), df_norm[col].max(), df_norm[col].mean(), df_norm[col].std()])

table = tabulate(rows, headers=header, tablefmt='pipe', floatfmt=".4f")
print(table)
```

```python
import pandas as pd
import numpy as np

# Leitura do arquivo parkinsons.data
df = pd.read_csv('parkinsons.data')

# Criação de DataFrames vazios para pacientes com status 0 e 1
df_aleatorio_0 = pd.DataFrame(columns=df.columns)
df_aleatorio_1 = pd.DataFrame(columns=df.columns)

# Loop para gerar 1200 pacientes com status 0 e 1200 pacientes com status 1
for i in range(1, 1201):
    # Criação de um novo DataFrame com valores aleatórios entre o mínimo e o máximo de cada coluna com status 0
    new_patient_0 = pd.DataFrame(columns=df.columns)
    new_patient_0['name'] = [f'paciente_aleatorio_{i}_0_np']
    new_patient_0['status'] = 0
    new_patient_1 = pd.DataFrame(columns=df.columns)
    new_patient_1['name'] = [f'paciente_aleatorio_{i}_1_np']
    new_patient_1['status'] = 1
    for col in df.columns[1:]:
        if col != 'status':
            min_val_0 = df.loc[df['status'] == 0][col].min()
            max_val_0 = df.loc[df['status'] == 0][col].max()
            min_val_1 = df.loc[df['status'] == 1][col].min()
            max_val_1 = df.loc[df['status'] == 1][col].max()
            new_val_0 = np.random.uniform(min_val_0, max_val_0)
            new_val_1 = np.random.uniform(min_val_1, max_val_1)
            if np.isnan(new_val_0):
                new_val_0 = 0
            if np.isnan(new_val_1):
                new_val_1 = 0
            new_patient_0[col] = [new_val_0]
            new_patient_1[col] = [new_val_1]
    df_aleatorio_0 = df_aleatorio_0.append(new_patient_0)
    df_aleatorio_1 = df_aleatorio_1.append(new_patient_1)

# Concatenação dos DataFrames com pacientes de status 0 e 1
df_aleatorio = pd.concat([df_aleatorio_0, df_aleatorio_1], ignore_index=True)

# Exportação do novo conjunto de dados
df_aleatorio = df_aleatorio.round(4)  # Arredondar valores para 4 casas decimais
df_aleatorio.to_csv('parkinsons_aleatorio_geral.csv', index=False)
```
