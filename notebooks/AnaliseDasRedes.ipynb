```python
import pandas as pd
import os
import tensorflow as tf
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Carrega os modelos salvos em formato .h5 da pasta
model_dir = 'modelos'
models = []
for model_file in os.listdir(model_dir):
    if model_file.endswith('.h5'):
        model_path = os.path.join(model_dir, model_file)
        model = tf.keras.models.load_model(model_path)
        models.append(model)

# Carrega o arquivo CSV com as amostras
amostras_df = pd.read_csv('amostras\parkinsons_aleatorio_1.csv')

# Normaliza os dados
scaler = MinMaxScaler()
amostras_norm = scaler.fit_transform(amostras_df)

# Testa cada linha do CSV nos modelos e retorna o status
statuses = []
for row_values in amostras_norm:
    row_values = row_values.reshape(1, 22, -1)
    predictions = []
    for model in models:
        print("Verificando :",model)
        prediction = model.predict(row_values.transpose())
        predictions.append(prediction[0][0])
    mean_prediction = np.mean(predictions)
    status = 1 if mean_prediction >= 0.5 else 0
    statuses.append(status)

# Após fazer a análise de todas as linhas, cria um DataFrame com os resultados
results_df = pd.DataFrame({'status': statuses})

# Salva o DataFrame em um arquivo CSV
results_df.to_csv('resultados.csv', index=False)
```
